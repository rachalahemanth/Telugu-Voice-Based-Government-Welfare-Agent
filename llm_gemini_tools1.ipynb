```python
class ConversationMemory:
    def __init__(self, max_turns: int | None = None):
        """
        max_turns: Optional limit on number of stored turns
        """
        self.history = []
        self.max_turns = max_turns

    # -----------------------------
    # Append methods (NO overwrite)
    # -----------------------------
    def add_user_message(self, text: str):
        self._append("user", text)

    def add_agent_message(self, text: str):
        self._append("agent", text)

    def _append(self, role: str, text: str):
        self.history.append({
            "role": role,
            "content": text
        })

        # Optional trimming (keep latest turns only)
        if self.max_turns is not None:
            self.history = self.history[-self.max_turns:]

    # -----------------------------
    # Read methods
    # -----------------------------
    def get_formatted_memory(self) -> str:
        """
        Formats memory into a readable string for LLM prompts
        """
        formatted = ""
        for turn in self.history:
            prefix = "‡∞µ‡∞ø‡∞®‡∞ø‡∞Ø‡±ã‡∞ó‡∞¶‡∞æ‡∞∞‡±Å" if turn["role"] == "user" else "‡∞è‡∞ú‡±Ü‡∞Ç‡∞ü‡±ç"
            formatted += f"{prefix}: {turn['content']}\n"
        return formatted.strip()

    def get_raw_memory(self) -> list:
        """
        Returns raw memory list (for debugging or logging)
        """
        return self.history.copy()

    # -----------------------------
    # Maintenance methods
    # -----------------------------
    def reset(self):
        """
        Clears all conversation history
        """
        self.history.clear()

    def size(self) -> int:
        """
        Returns number of stored turns
        """
        return len(self.history)
```

```python
def agent_loop(user_text: str, memory: ConversationMemory):
    # Save user input
    memory.add_user_message(user_text)

    # Planner decides tool sequence
    plan = planner_agent(user_text)

    # Execute tools
    tool_outputs = execute_steps(plan["steps"])

    # Generate final answer using memory
    final_answer = generate_answer_with_memory(
        user_text,
        tool_outputs,
        memory
    )

    # Save agent response
    memory.add_agent_message(final_answer)

    return final_answer.replace("*", "")

```
